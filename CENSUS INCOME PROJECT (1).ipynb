```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tools.eval_measures import mse,rmse
import statsmodels.api as sm
!pip install pmdarima
import pmdarima as pm
```

```python
#as we are seeing abovve our data is not stationary by visualizing the seasonal decompose we are using first rolling stats to make our data stationary
rolling_mean=data - data.shift(24)
rolling_mean.plot(figsize=(12,5),grid=True)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decompose_result = seasonal_decompose(rolling_mean.dropna())
#We can check the trend and seasonality using the
# seasonal_decompose() method.
decompose_result.plot();


#Here in the  chart, we can see the decomposed
# structure of data and the structure of the components 
#in the data set which were affecting it.
#A- as we can see the constant trend and seasonality in every year of the data
```

```python
rolling_mean_na=rolling_mean.dropna()
```

```python
def adf_test(series):
    result = adfuller(series)
    print('ADF Statistics : {}', format(result[0]))
    print('p-value : {}', format(result[1]))
    if result[1] <= 0.05:
        print('Strong evidence against the Null Hypothesis, rejecting Null Hypothesis. Data has no unit root and is Stationary')
    else:
        print('Weak evidence against the Null Hypothesis, time series has a unit root, indicating it is Non-Stationary')
```

```python
adf_test(rolling_mean_na['value'])
#as we can see our pvalue is less than 0.05 so it means our data is stationary now
```

```python
plot_acf(rolling_mean_na['value']);
```

```python
plot_pacf(rolling_mean_na['value']);
```

```python
model_auto_arima = pm.auto_arima(rolling_mean_na['value'], 
                                 start_p = 1, 
                                 start_q = 1, 
                                 test = 'adf',
                                 max_p=5, 
                                 max_q=5, 
                                 seasonal = False, 
                                 start_P=0, 
                                 D=0, 
                                 trace=True,
                                 error_action='ignore',  
                                 suppress_warnings=True, 
                                 stepwise=True)
print(model_auto_arima.summary())
```

```python
train = rolling_mean_na.iloc[:602]['value']
test = rolling_mean_na.iloc[603:]['value']

model = ARIMA(train, order=(3,1,3))
model_fit = model.fit()
model_fit.summary()
```

```python
rolling_mean_na['predict'] = model_fit.predict(start= len(train), 
                                    end=len(train)+len(test)- 1, 
                                    dynamic=True)
rolling_mean_na[['value','predict']].plot()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX, SARIMAXResults

model = SARIMAX(train, order=(3,1,3), seasonal_order=(3,1,3,12))
model = model.fit()
```

```python
rolling_mean_na['predict'] = model.predict(start= len(train) ,
                                                       end=len(train)+len(test)- 1, 
                                                       dynamic=True)
rolling_mean_na[['value','predict']].plot()
```
